```python
import numpy as np

class NormalizedPenaltyCalculator:
    def __init__(self, base_weight=0.5, threshold=0.2):
        """
        Initialize penalty calculator for 0-1 range values.
        
        Args:
            base_weight (float): Base penalty weight (default: 0.5)
            threshold (float): Threshold for penalty transitions (default: 0.2)
                             Smaller threshold = more sensitive to small differences
        """
        self.base_weight = base_weight
        self.threshold = threshold
        
    def calculate_penalty(self, x_target, x_current, mode="adaptive"):
        """
        Calculate penalty for values in 0-1 range.
        
        Args:
            x_target (float): Target value (0-1)
            x_current (float): Current value (0-1)
            mode (str): Penalty mode: "adaptive", "sigmoid", "quadratic", or "smooth"
        
        Returns:
            float: Penalty value (0-1)
        """
        # Ensure inputs are in valid range
        x_target = np.clip(x_target, 0, 1)
        x_current = np.clip(x_current, 0, 1)
        difference = abs(x_target - x_current)
        
        if mode == "adaptive":
            if difference < self.threshold:
                # Quadratic for small differences - more sensitive
                penalty = (difference / self.threshold) ** 2
            else:
                # Smooth linear for larger differences
                penalty = difference / (1 - self.threshold)
                
        elif mode == "sigmoid":
            # Smooth S-shaped penalty
            k = 10  # steepness
            midpoint = self.threshold
            penalty = 1 / (1 + np.exp(-k * (difference - midpoint)))
            
        elif mode == "quadratic":
            # Pure quadratic penalty - sensitive to small differences
            penalty = (difference) ** 2
            
        elif mode == "smooth":
            # Smooth polynomial transition
            penalty = difference * (3 - 2 * difference)
            
        else:
            raise ValueError(f"Unknown mode: {mode}")
            
        # Apply base weight and ensure output range
        return np.clip(round(self.base_weight * penalty, 3), 0, 1)
    
    def adaptive_weight_update(self, episode_performance):
        """
        Update base_weight based on episode performance (0-1 scale)
        """
        self.base_weight = np.clip(
            self.base_weight * (1 + 0.1 * (0.5 - episode_performance)),
            0.1,  # minimum weight
            0.9   # maximum weight
        )

# Example usage
if __name__ == "__main__":
    calculator = NormalizedPenaltyCalculator(base_weight=0.5, threshold=0.2)
    
    # Test with values in 0-1 range
    target = 0.849
    current = 0.85
    
    print(f"Target: {target}, Current: {current}, Difference: {abs(target-current)}")
    print(f"Adaptive penalty: {calculator.calculate_penalty(target, current, 'adaptive')}")
    print(f"Sigmoid penalty: {calculator.calculate_penalty(target, current, 'sigmoid')}")
    print(f"Quadratic penalty: {calculator.calculate_penalty(target, current, 'quadratic')}")
    print(f"Smooth penalty: {calculator.calculate_penalty(target, current, 'smooth')}")
```

```python
from helpers.Constants import UNIVERSAL_POSSIBLE_PERCENTAGES

def get_possible_actions(heads):
    """
    This function returns all possible combinations of resource allocations for all heads.
    Each head can have different combinations of [bitrate, framerate, behavioral_accuracy].
    The bitrate, framerate and behavioral_accuracy are given as percentages of the maximum values.
    Returns:
        List[List[Tuple]]: List of all possible combinations for all heads
        int: Total number of possible actions
    """

    def generate_head_combinations(head_index, current_combination):
        # Base case: if we've assigned resources to all heads
        if head_index == len(heads):
            return [current_combination[:]]

        combinations = []
        # For each head, try all possible combinations of resources
        for max_bit_rate_perc in UNIVERSAL_POSSIBLE_PERCENTAGES:
            for max_frame_rate_perc in UNIVERSAL_POSSIBLE_PERCENTAGES:
                for max_behav_acc_perc in UNIVERSAL_POSSIBLE_PERCENTAGES:
                    # Append the current combination for this head
                    current_combination.append((max_bit_rate_perc, max_frame_rate_perc, max_behav_acc_perc))
                    # Recurse to generate combinations for the next head
                    combinations.extend(generate_head_combinations(head_index + 1, current_combination))
                    # Remove the last added combination to backtrack
                    current_combination.pop()

        return combinations

    # Generate all possible combinations starting with an empty list
    actions_dict = generate_head_combinations(0, [])
    number_of_actions = len(actions_dict)
    # self.actions_generated = True
    # print(f"@{self.__class__.__name__}, Info: MSP {self.id} has generated {number_of_actions} actions")
    my_dict = {index: value for index, value in enumerate(actions_dict)}
    return my_dict, number_of_actions
```

```python
def get_possible_help_actions(neighbors=[1,1]):
    """
    This function will create possible help actions for the msp. 
    sample call: get_possible_actions([1])[0]
    sample return is {0: 0.2, 1: 0.4, 2: 0.6, 3: 0.8, 4: 1.0}
    """
    def generate_head_combinations(neighor_index, current_combination):
                # Base case: if we've assigned resources to all heads
                if neighor_index == len(neighbors):
                    return [current_combination[:]]

                combinations = []
                # For each head, try all possible combinations of resources
                for max_bit_rate_perc in np.arange(0, 1.2, 0.2):
                    # current_combination.append((max_bit_rate_perc))
                    current_combination.append(max_bit_rate_perc)
                    # Recurse to generate combinations for the next head
                    combinations.extend(generate_head_combinations(neighor_index + 1, current_combination))
                    # Remove the last added combination to backtrack
                    current_combination.pop()
                return combinations
    
    actions_dict = generate_head_combinations(0, [])
    number_of_actions = len(actions_dict)
    # self.actions_generated = True
    action_dict = {index: value[0] for index, value in enumerate(actions_dict)}
    return action_dict, number_of_actions
```

```python
def help_value_to_discrete(value,max_value):
    """
    This function takes a value from -1 to 1 and outputs a discrete number from 0 to 4 inclusive.
    Each discrete value represents an equal-width bin in the continuous space.
    """
    assert -1 <= value <= 1, "Value must be between -1 and 1 inclusive."
    normalized_value = (value + 1) / 2
    discrete_value = min(int(normalized_value * max_value), max_value-1)
    return discrete_value
```

```python
def get_possible_help_actions(num_neighbors=1):
    """
    Generate all possible combinations of help values for neighbors
    Each neighbor can have help values from 0.1 to 0.9 in steps of 0.1
    Returns a list of lists where each inner list contains dictionaries with index and help values
    """
    combinations = []
    help_values = np.arange(0, 0.8, 0.2)
    
    # Generate all possible combinations of help values for each neighbor
    for help_values_combination in itertools.product(help_values, repeat=num_neighbors):
        combination = []
        for help_value in help_values_combination:
            # combination.append({"index": index, "help": help_value})
            combination.append(help_value)
        combinations.append(combination)
    
    return dict(enumerate(combinations)),len(combinations)
```
